````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# data read

demograhpic_data = pd.read_csv("data/life_expectancy_and_income.csv")
````

````python
demographic_data_2019 = (
    demograhpic_data
    .loc[demograhpic_data.year == 2019]
    .copy()
)
````

For the year 2019, how did life expectancy compare to income for each country?

````python
plt.scatter(x="income_per_person",
           y="life_expectancy",
           data=demographic_data_2019
)
````

````python
# base plot
plt.scatter(x="income_per_person",
           y="life_expectancy",
           data=demographic_data_2019
)

# axis labels
plt.xlabel("Income per person ($)")
plt.ylabel("Life expectancy (years)")

# plot title
plt.title("Life Expectancy vs Income - 2019")
````

````python
# base plot
plt.scatter(x="income_per_person",
           y="life_expectancy",
           data=demographic_data_2019
)

# axis labels
plt.xlabel("Income per person ($)")
plt.ylabel("Life expectancy (years)")

# plot title
plt.title("Life Expectancy vs Income - 2019")

# set limits (this plot is probably fine)
plt.xlim(0, 80000)
plt.ylim(0, 100)
````

````python
# base plot
plt.scatter(x="income_per_person",
           y="life_expectancy",
           data=demographic_data_2019
)

# axis labels
plt.xlabel("Income per person ($)")
plt.ylabel("Life expectancy (years)")

# plot title
plt.title("Life Expectancy vs Income - 2019")

# set limits (this plot is probably fine)
plt.xlim(0, 80000)
plt.ylim(0, 100)

# control tick marks
plt.yticks(list(range(0,110,10)))

# only shows plot, no rendering code
plt.show
````

### Histograms

* same plotting options
* different `plt.plot_type`

````python
plt.hist(x="income_per_person",
        data=demographic_data_2019)

plt.xlabel("Income per person ($)")
plt.ylabel("Count")
plt.title("Distribution of income per person for 2019")

plt.show()
````

````python
plt.boxplot(x="income_per_person",
           data=demographic_data_2019)

plt.xlabel("2019")
plt.ylabel("Income per person ($)")
plt.title("Distribution of income per person for 2019")

plt.show()
````

````python
occupation_prestige = pd.read_csv("data/prestige_occupation.csv")
````

````python
sns.scatterplot(x="prestige",
               y="education",
               data=occupation_prestige)

plt.xlabel("Prestige score")
plt.ylabel("Education score")
plt.title("Prestige vs Education")

plt.show()
````

````python
min_score = (
    occupation_prestige.loc[
        occupation_prestige
        .groupby("type")
        .prestige
        .idxmin()
        .sort_values()]
    .copy()
)
````

````python
sns.barplot(x="type",
           y="prestige",
           data=min_score,
           color="steelblue")

plt.xlabel("Type")
plt.ylabel("Prestige score")
plt.title("Minimum Prestige Score per Type")

plt.show()
````

Categorical colours

What if we want to draw attention to particularly countries where fertility rates is high?
<br>
Here, we will need to do some pre-plot wrangling (make a high_fertility_rate column)
<br>
high_fertility_rate when fertility_rate > 5

````python
import numpy as np
````

```python
np.where(condition, value_if_true, value_is_false)
```

````python
demographic_data_2019["high_fertility_rate"] = np.where(
    demographic_data_2019.fertility_rate > 5, True, False
)

demographic_data_2019.head()
````

````python
sns.scatterplot(x="income_per_person",
                y="life_expectancy",
                data=demographic_data_2019,
                hue="high_fertility_rate",
                palette="dark"
)
````

````python
# apply own palette and change point style

sns.scatterplot(x="income_per_person",
                y="life_expectancy",
                data=demographic_data_2019,
                hue="high_fertility_rate",
                palette={True: "#fc0313", False: "#030bfc"},
                style="high_fertility_rate"
)
````
